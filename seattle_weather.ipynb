```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import pickle
```

```python
weather_df = pd.read_csv('seattle_weather.csv')
weather_df.head()
```

```python
X = weather_df.drop(['location','date','weather'], axis=1)
y = weather_df['weather']
xData = X.values
yData = y.values
```

```python

xTrain, xTest, yTrain, yTest = train_test_split(xData, yData, test_size = 0.2, random_state = 42)
```

```python

# random forest model creation
rfc = RandomForestClassifier(n_estimators=100, random_state=42)
rfc.fit(xTrain, yTrain)
# predictions
yPred = rfc.predict(xTest)
```

```python
accuracy_score = rfc.score(xTest, yTest)
print(accuracy_score)
```

```python
from sklearn.metrics import classification_report
print(classification_report(yTest, yPred))
```

```python
test = [0.0, 12.8 ,5.0, 4.7]
print(rfc.predict([test]))
```

```python
with open('model.pkl', 'wb') as f:
    pickle.dump(rfc, f)
```

```python
import numpy as np
from scipy.stats import linregress

def temperature_trend(temperatures):
    hours = np.arange(len(temperatures))
    slope, intercept, r_value, p_value, std_err = linregress(hours, temperatures)
    
    return slope

# Ví dụ sử dụng
temperatures = [24.0, 24.1, 24.8, 24.4, 24.0, 24.4, 24.8, 25.0, 26.0, 26.8, 27.3, 27.1, 28.2, 29.1, 29.4, 28.1, 28.0, 27.9, 27.5, 27.4, 26.9, 26.0, 25.2, 24.7]
trend = temperature_trend(temperatures)
print(f"Xu hướng nhiệt độ: {trend}")
```
